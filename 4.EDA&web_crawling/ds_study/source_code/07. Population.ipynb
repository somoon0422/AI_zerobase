```python
population = pd.read_excel(
    '../data/07_population_raw_data.xlsx', header=1
)

population.fillna(method='pad', inplace=True) # 결측치 처리 pad는 앞의 값으로 채움
```

```python
# 컬럼 이름 변경

population.rename(
    columns={
        "행정구역(동읍면)별(1)" :"광역시도",
        "행정구역(동읍면)별(2)" :"시도",
        '계': "인구수"
    }, inplace=True
)

population.tail()
```

```python
# 소계 제거

population = population[population["시도"] != "소계"]
```

```python
population.head()
```

```python
population.is_copy = False # copy warning 제거

population.rename(
    columns={
        "항목" : "구분"
    }, inplace=True
)
```

```python
population.head()
```

```python
population.loc[population['구분']=="총인구수 (명)","구분"] = "합계"
population.loc[population['구분']=="남자인구수 (명)","구분"] = "남자"
population.loc[population['구분']=="여자인구수 (명)","구분"] = "여자"
```

```python
population.head()
```

```python
population["20~39세"] = (
    population["20 - 24세"] +
    population["25 - 29세"] +
    population["30 - 34세"] +
    population["35 - 39세"]
)

population["65세이상"] = (
    population["65 - 69세"] +
    population["70 - 74세"] +
    population["75 - 79세"] +
    population["80 - 84세"] +
    population["85 - 89세"] +
    population["90 - 94세"] +
    population["95 - 99세"] +
    population["100+"]
)
```

```python
population.head()
```

```python
# pivot_table

pop = pd.pivot_table(
    data=population,
    index=["광역시도","시도"],
    columns=["구분"],
    values=["인구수", "20~39세", "65세이상"]
)

pop
```

```python
# 소멸 비율 계산

pop['소멸비율'] = pop["20~39세","여자"] / (pop["65세이상", "합계"] / 2)
pop.tail()
```

```python
# 소멸 위기 지역 컬럼 생성

pop['소멸위기지역'] = pop["소멸비율"] < 1.0 # True, False로 나옴
pop.head()
```

```python
## 소멸위기지역 조회

pop[pop["소멸위기지역"] == True].index.get_level_values(1)
```

#### (1) 일반 시 이름과 세종시,광역시도 일반 구 정리

```python
for idx, row in pop.iterrows():
    if row["광역시도"][-3] not in ["광역시", "특별시","자치시"]:
        si_name[idx] = row["시도"][:-1]
        
    elif row['광역시도'] == "세종특별자치시" :
        si_name[idx] = "세종"
    
    else:
        if len(row["시도"]) == 2:
            si_name[idx] = row["광역시도"][:2] + " " + row["시도"] # ex) 서울특별시 서초구 -> 서울 서초구
        else:
            si_name[idx] = row["광역시도"][:2] + " " + row["시도"][:-1] # ex) 서울특별시 동작구 -> 서울 동작구
```

### (2) 행정구

```python
for idx, row in pop.iterrows():
    if row["광역시도"][-3:] not in ["광역시", "특별시","자치시"]:
        for keys, values in tmp_gu_dict.items():
            if row["시도"] in values:
                if len(row["시도"]) == 2:
                    si_name[idx] = keys + " " + row["시도"]
                    
                elif row["시도"] in ["마산합포구", "마산회원구"]:
                    si_name[idx] = keys + " " + row["시도"][2:-1]
                    
                else:
                    si_name[idx] = keys + " " + row["시도"][:-1]
```

```python
si_name
```

#### (3) 고성군

```python
for idx, row in pop.iterrows():
    if row["광역시도"][-3:] not in ["광역시","특별시","자치시"]:
        if row["시도"][:-1] == "고성" and row["광역시도"] == "강원도":
            si_name[idx] = "고성(강원)"
        elif row["시도"][:-1] == "고성" and row["광역시도"] == "경상남도":
            si_name[idx] = "고성(경남)"
```

```python
BORDER_LINES = [
    [(5,1),(5,2),(7,2),(7,3),(11,3),(11,0)], # 인천
    [(5,4),(5,5),(2,5),(2,7),(4,7),(4,9),(7,9),(7,7),(9,7),(9,5),(10,5),(10,4)], # 서울
    [(1,7),(1,8),(3,8),(3,10),(10,10),(10,7),(12,7),(12,6),(11,6),(11,5),(12,5),(12,4),(11,4),(11,3)], # 경기도
    [(8,10),(8,11),(6,11),(6,12)], # 강원도
    [(12,5),(13,5),(13,4),(14,4),(14,5),(15,5),(15,4),(16,4),(16,2)], # 충청북도
    [(16,4),(17,4),(17,5),(16,5),(16,6),(19,6),(19,5),(20,5),(20,4),(21,4),(21,3),(19,3),(19,1)], # 전라북도
    [(13,5),(13,6),(16,6)], # 대전시
    [(13,5),(14,5)], #세종시
    [(21,2),(21,3),(22,3),(22,4),(24,4),(24,2),(21,2)], # 광주
    [(20,5),(21,5),(21,6),(23,6)], # 전라남도
    [(10,8),(12,8),(12,9),(14,9),(14,8),(16,8),(16,6)], # 충청북도
    [(14,9),(14,11),(14,12),(13,12),(13,13)], # 경상북도
    [(15,8),(17,8),(17,10),(16,10),(16,11),(14,11)], # 대구
    [(17,9),(18,9),(18,8),(19,8),(19,9),(20,9),(20,10),(21,10)], # 부산
    [(16,11),(16,13)], # 울산
]
```

```python
BORDER_LINES
```

```python
def plot_text_simple(draw_korea):
    for idx, row in draw_korea.iterrows():
        if len(row["ID"].split()) == 2:
            dispname = "{}\n{}".format(row["ID"].split()[0], row["ID"].split()[1])
        elif row["ID"][:2] == "고성":
            dispname = "고성"
        else:
            dispname = row['ID']
    
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 9.5, 1.5
        else:
            fontsize, linespacing = 11, 1.2
            
        # 주석 다는 기능
        plt.annotate( # annotate는 주석을 다는 기능
            dispname,
            (row["x"]+0.5, row["y"]+0.5),
            weight="bold",
            fontsize = fontsize,
            linespacing = linespacing,
            ha = "center", #수평 정렬
            va = "center" # 수직 정렬
    )
```

```python
def simpleDraw(draw_korea):
    plt.figure(figsize=(8,11))
    
    plot_text_simple(draw_korea)
    
    for path in BORDER_LINES: # 경계선 그리기
        ys, xs = zip(*path) # zip은 묶어주는 역할
        plt.plot(xs , ys, c='black', lw=1.5) # c는 색깔, lw는 선의 굵기 , *은 unpacking
        
    plt.gca().invert_yaxis() # y축을 뒤집어 줌
    plt.axis("off") # 축을 없애줌
    plt.tight_layout() # 그래프를 자동으로 맞춰줌
    plt.show() # 그래프를 보여줌
```

#### 그림을 그리기 위한 데이터를 계산하는 함수
- 색상을 만들 떄, 최소값을 흰색
- blockedMap : 인구현황(pop)
- targetData : 그리고 싶은 컬럼

```python
def get_data_info(targetData, blockedMap):
    whiteabelmin = (
        max(blockedMap[targetData]) - min(blockedMap[targetData])
    ) * 0.25 + min(blockedMap[targetData])
    
    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])
    
    mapdata = blockedMap.pivot_table(
        index="x", columns="y", values=targetData
    )
    
    return mapdata, vmax, vmin, whiteabelmin
```

```python
def get_data_info_for_zero_center(targetData, blockedMap):
    whiteabelmin =5
    tmp_max = max(
        [np.abs(min(blockedMap[targetData])), np.abs(max(blockedMap[targetData]))]
    )
    
    vmin , vmax = -tmp_max, tmp_max
    mapdata = blockedMap.pivot_table(
        index="y", columns="x", values=targetData # pivot_table은 행과 열을 바꿔줌
    )
    
    return mapdata, vmax, vmin, whiteabelmin
```

```python
def plot_text(targetData, blockedMap, whiteabelmin):
    for idx, row in draw_korea.iterrows():
        if len(row["ID"].split()) == 2:
            dispname = "{}\n{}".format(row["ID"].split()[0], row["ID"].split()[1])
        elif row["ID"][:2] == "고성":
            dispname = "고성"
        else:
            dispname = row['ID']
    
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 9.5, 1.5
        else:
            fontsize, linespacing = 11, 1.2
            
        annocolor = "white" if np.abs(row[targetData]) > whiteabelmin else "black"
            
        # 주석 다는 기능
        plt.annotate( # annotate는 주석을 다는 기능
            dispname,
            (row["x"]+0.5, row["y"]+0.5),
            weight="bold",
            color = annocolor,
            fontsize = fontsize,
            linespacing = linespacing,
            ha = "center", #수평 정렬
            va = "center" # 수직 정렬
    )
```

```python
def drawkorea(targetData, blockedMap, zeroCenter=False):
    
    if zeroCenter:
        masked_mapdata, vmax, vmin, whiteabelmin = get_data_info_for_zero_center(targetData, blockedMap)
    
    if not zeroCenter:
        masked_mapdata, vmax, vmin, whiteabelmin = get_data_info(targetData, blockedMap)
        
    plt.figure(figsize=(8,11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap="RdBu_r", edgecolor="#aaaaaa", linewidth=0.5)
    
    plot_text(targetData, blockedMap, whiteabelmin)
    
    for path in BORDER_LINES: # 경계선 그리기
        ys, xs = zip(*path) # zip은 묶어주는 역할
        plt.plot(xs , ys, c='black', lw=1.5) # c는 색깔, lw는 선의 굵기 , *은 unpacking
        
    plt.gca().invert_yaxis() # y축을 뒤집어 줌
    plt.axis("off") # 축을 없애줌
    plt.tight_layout() # 그래프를 자동으로 맞춰줌
    cb = plt.colorbar(shrink=0.1, aspect=10)
    cb.set_label(targetData)
    plt.show() # 그래프를 보여줌
```

```python
drawkorea("인구수합계", pop, "Blues")
```
