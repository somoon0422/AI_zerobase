````python
import requests
#from urllib.request import urlopen
from bs4 import BeautifulSoup
````

````python
url = "https://finance.naver.com/marketindex/"
response = requests.get(url) 
# resuqests.get()
# requests.post()
#  url을 통해서 서버에 요청
#response # 200 : 정상 (HTTP상태코드)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
````

````python
#class 면 .을 사용 하고, id 면 #을 사용

exchangeList = soup.select("#exchangeList > li") 
len(exchangeList) , exchangeList[0]
````

````python
title = exchangeList[0].select_one(".h_lst").text
exchange = exchangeList[0].select_one(".value").text
change = exchangeList[0].select_one(".change").text
updown = exchangeList[0].select_one("div.head_info.point_dn > .blind").text



title, exchange , change , updown
````

````python
findmethod = soup.find_all('ul', id="exchangeList")
findmethod[0].find_all('span', 'value')
````

````python
baseurl = "https://finance.naver.com"

link = baseurl + exchangeList[0].select_one('a').get('href')
link
````

````python
# 4개데이터 수집
import pandas as pd
exchange_datas = []
baseurl = "https://finance.naver.com"

for item in exchangeList:
    data = {
        "title" : item.select_one(".h_lst").text,
        "exchange" : item.select_one(".value").text,
        "change" : item.select_one(".change").text,
        "updown" : item.select_one("div.head_info.point_dn > .blind").text,
        "link" : baseurl + item.select_one('a').get('href')
    } 
    
    exchange_datas.append(data)
    
df = pd.DataFrame(exchange_datas)
df.to_excel('./naverfinance.xlsx' , index=False) # index=False : 인덱스 제거
````

#### 4. 시카고 맛집 데이터 분석 - 하위페이지

````python
# requirments

import pandas as pd
from urllib.request import urlopen , Request
from fake_useragent import UserAgent
from bs4 import BeautifulSoup

````

````python
df = pd.read_csv('../data/03. best_sandwiches_list_chicago.csv', index_col=0)
df.tail()
````

````python
df['URL'][0]
````

````python
req = Request(df['URL'][0], headers={"user-agent":UserAgent().chrome})
html = urlopen(req)
soup_tmp = BeautifulSoup(html, 'html.parser')
soup_tmp.find("p", "addy")
````

````python
# regular expreession

price_tmp = soup_tmp.find("p", "addy").text
price_tmp
````

````python
import re
re.split(".,",price_tmp)
````

````python
price_tmp = re.split(".,",price_tmp)[0]
price_tmp
````

````python
tmp = re.search("\$\d+\.(\d+)?", price_tmp).group()
price_tmp[len(tmp)+2:]
````

````python
from tqdm import tqdm

price = []
address = []

for idx, row in tqdm(df.iterrows()):
    req = Request(row['URL'], headers={"user-agent":UserAgent().chrome})
    html = urlopen(req)
    soup_tmp = BeautifulSoup(html, 'html.parser')
    
    gettings = soup_tmp.find("p", "addy").get_text()
    price_tmp = re.split(",.", gettings)[0]
    tmp = re.search("\$\d+\.(\d+)?", price_tmp).group()
    price.append(tmp)
    address.append(price_tmp[len(tmp)+2:])
    print(idx)
````

````python
len(price), len(address)
````

````python
df.tail(2)
````

````python
df['Price'] = price
df['Address'] = address
df = df.loc[:, ["Rank","Cafe","Menu","Price","Address"]]
df.set_index("Rank", inplace = True)
df.head()
````

````python
df.to_csv(
    "../data/03. best_sandwiches_list_chicago2.csv",
    sep=',',
    encoding='UTF-8'
)
````

#### 5. 시카고 맛집 데이터 지도 시각화

````python
# requirments

import folium
import pandas as pd
import numpy as np
import googlemaps
from tqdm import tqdm
````

````python
df = pd.read_csv("../data/03. best_sandwiches_list_chicago2.csv", index_col=0)
df.tail(10)
````

````python
gmaps_key = "AIzaSyB_8YZKACITy1OGyZT8hrIR_y4W3TpfXGM"
gmaps = googlemaps.Client(key=gmaps_key)
````

````python
lat = []
lng = []

for idx, row in tqdm(df.iterrows()):
    if not row['Address'] == "Multiple location":
        target_name = row['Address'] + ", " + "Chicago"
        # print(target_name)
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get("geometry")
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
        # location_output = gmaps.output[0]
        
    else:
        lat.append(np.nan)
        lng.append(np.nan)
        
````

````python
len(lat), len(lng)
````

````python
df['lat'] = lat
df['lng'] = lng
df.tail()
 
````

````python
mapping = folium.Map(location=[41.8781136, -87.6297982], zoom_start=11)

for idx, row in df.iterrows():
   if not row['Address'] == "Multiplt location":
       folium.Marker(
           location=[row['lat'], row['lng']],
           popup=row['Cafe'],
           tooltip=row["Menu"],
           icon=folium.Icon(
               icon="coffee",
               prefix="fa"
           )
       ).add_to(mapping)
       
mapping       
````
