```python
# Decision Tree 분할기준
# split criterion : gini, entropy
# gini : 불순도를 나타내는 지표
# entropy : 불확실성을 나타내는 지표
# entropy = -sigma(pi * log(pi))
# entropy = 0 : 모든 데이터가 같은 클래스
# entropy = 1 : 클래스가 2개
# entropy = 2 : 클래스가 3개
# gini = 0 : 모든 데이터가 같은 클래스
# gini = 1 : 클래스가 2개
# gini = 2 : 클래스가 3개
# gini = entropy * 2

# 지니계수는 낮을수록 좋은 분할
# 에트로피도 낮을수록 좋은 분할


# 정보 획득 : information gain
# 정보 획득이란? : 부모 노드와 자식 노드의 불순도 차이
# 정보 획득이 클수록 좋은 분할
# 열역학 제 2법칙 : 엔트로피는 항상 증가한다.
```

```python
import numpy as np

p = np.arange(0.001, 1, 0.001)
plt.grid()
plt.title('$-p \log_{2}{p}$')
plt.plot(p, -np.log2(p), label='log2')
```

# Scikit Learn

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier()
```

```python
iris.data
```

```python
iris_tree.fit(iris.data[:,2:], iris.target ) # petal length, petal width

# iris.target: 0, 1, 2 
```

```python
from sklearn.metrics import accuracy_score # 정확도 측정

y_pred_tr = iris_tree.predict(iris.data[:, 2:]) # 예측
y_pred_tr
```

```python
accuracy_score(iris.target, y_pred_tr) # 정확도 측정
```

```python
from sklearn.tree import plot_tree # 트리 그리기

plt.figure(figsize=(12,8))

plot_tree(iris_tree)
```
