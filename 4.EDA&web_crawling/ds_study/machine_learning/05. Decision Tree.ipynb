```python
# 데이터 받아오기 

import pandas as pd

red_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/winequality-red.csv'

white_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/winequality-white.csv'

red_wine = pd.read_csv(red_url, sep=';')
white_wine = pd.read_csv(white_url, sep=';')

```

```python
red_wine.head()
```

```python
white_wine.head()
```

```python
white_wine.columns
```

```python
red_wine['color'] = 1.
white_wine['color'] = 0.

wine = pd.concat([red_wine, white_wine])
wine.info()
```

```python
wine.head()
```

```python
wine['quality'].unique()
```

```python
# 시각화

import plotly.express as px

flg = px.histogram(wine, x='quality', color='color')
flg
```

```python
# 라벨분리

X = wine.drop(['color'], axis=1)
y = wine['color']

```

```python
# 훈련용과 테스트용으로 나누기

from sklearn.model_selection import train_test_split
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
np.unique(y_train,return_counts=True)
```

```python
# 훈련/테스트용 데이터가 레드와인과 화이트 와에 따라 어느정도 구분되었을까?

import plotly.graph_objects as go

flg = go.Figure()
flg.add_trace(go.Histogram(x=X_train['quality'], name='Train'))
flg.add_trace(go.Histogram(x=X_test['quality'], name='Test'))

flg.update_layout(barmode='overlay')
flg.update_traces(opacity=0.7)

flg.show()
```

```python
# 결정나무 훈련

from sklearn.tree import DecisionTreeClassifier

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
wine_tree.fit(X_train, y_train)


```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)
```

```python
def px_box(target_df):
    flg = go.Figure()
    flg.add_trace(go.Box(y=target_df['fixed acidity'], name='fixed acidity'))
    flg.add_trace(go.Box(y=target_df['chlorides'], name='chlorides'))
    flg.add_trace(go.Box(y=target_df['quality'], name='quality'))

    flg.show()
```

```python
wine['taste'] = [1. if grade > 5 else 0. for grade in wine['quality']]
wine.head()
```

```python
wine.info()
```

```python
X = wine.drop(['taste'], axis=1)
y = wine['taste']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
wine_tree.fit(X_train, y_train)
```

```python
y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print("Train Acc: " ,accuracy_score(y_train, y_pred_tr))
print("Test Acc: " ,accuracy_score(y_test, y_pred_test))
```

```python
import matplotlib.pyplot as plt 
import sklearn.tree as tree

plt.figure(figsize=(12,8))
tree.plot_tree(wine_tree, feature_names=X.columns)
```
